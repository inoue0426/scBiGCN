# Clustering

## Read data

```python
df = pd.read_csv('sample_data/sample_data.csv.gz', index_col=0)
df
```

## Preprocess for clustering

```python
tmp = np.sign(df)
cols = (np.sum(tmp) > int((df.shape[0])*0.05))
rows = (np.sum(tmp, axis=1) > int((df.shape[1])*0.05))
df = np.log(df.loc[rows, cols] + 1)
df = scprep.normalize.library_size_normalize(df)    
df = scprep.transform.sqrt(df)
df
```

## Run model

```python
%%time
imputed_matrix = bigcn.run_model(df, clustering=True, verbose=True)
imputed_matrix
```
